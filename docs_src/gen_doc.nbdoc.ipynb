`nbdoc` contains the functions for documentation notebooks. The most important is [`show_doc`](/gen_doc.nbdoc.html#show_doc):

````python
show_doc(get_source_link)
````

````python
show_doc(show_doc)
````

In your notebook you'll have a call to [`show_doc`](/gen_doc.nbdoc.html#show_doc) for every symbol you wish to document. And you'll want to *hide input* for that cell. For instance, the above signature is generated by this function:

```python
show_doc(show_doc)
```
